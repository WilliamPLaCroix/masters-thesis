```python
import json

variations = ["original", "cleaned", "augmented"]
grades = [f"grade{g:02d}" for g in range(2, 13)] + ["baseline"]
splits = ["train", "validation", "test"]

HF_REPO = "williamplacroix/graded_wikilarge"
COLUMNS = {"prompt": "instruction", "query": "input", "response": "output"}

def key(variation, grade, split):
    return f"{variation}_{grade}_{split}"

def entry(variation, grade, split):
    return {
        "hf_hub_url": HF_REPO,
        "folder": f"{variation}/{grade}",
        "split": split,
        "columns": COLUMNS
    }

data = {key(v, g, s): entry(v, g, s) for v in variations for g in grades for s in splits}

for k, v in data.items():
    print(f'"{k}": {v},')
with open("dataset_info.json", "w") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)
```

```python
# create folders within ./comparative_datasets/ for each variation/grade combination
import os

variations = ["original", "cleaned", "augmented"]
grades = [f"grade{g:02d}" for g in range(2, 13)] + ["baseline"]
splits = ["train", "validation", "test"]

for variation in variations:
    for grade in grades:
        folder_path = os.path.join("comparative_datasets", variation, grade)
        print(folder_path)
        os.makedirs(folder_path, exist_ok=True)
```

```python
""" Upload in one go: """

!hf auth login

from huggingface_hub import HfApi, create_repo, upload_folder
api = HfApi()

repo_id = "williamplacroix/graded_wikilarge"
api.create_repo(repo_id=repo_id, repo_type="dataset", exist_ok=True)

# Upload the whole tree from your local "data" dir into repo root
upload_folder(
    repo_id=repo_id,
    repo_type="dataset",
    local_dir="comparative_datasets",  # folder containing all variations
    path_in_repo="",                 # keep same structure
    allow_patterns=["**/*.json"],
    commit_message="Initial upload of 3 variations × 11 grades (+baseline) × 3 splits",
)


```
